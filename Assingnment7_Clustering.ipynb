 NOW THE K - CLUSTERING


Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdsclar=StandardScaler()
```

```python
scaled_crime=stdsclar.fit_transform(crime)
scaled_crime
```

BUILDING THE MODEL

```python
from sklearn.cluster import KMeans
```

```python
Kmodel=KMeans(n_clusters=4,random_state=42)
```

```python
Kmodel.fit(scaled_crime)
```

```python
Kmodel.labels_
```

Hyperparameter Tuning

```python
wcss_cd=[]
for i in range(1,12):
  model=KMeans(n_clusters=i,random_state=42)
  model.fit(crime)
  wcss_cd.append(model.inertia_)
  wcss_cd
```

```python
plt.plot(range(1,12),wcss_cd,color='Red')
plt.title('Elbow Plot')
plt.xlabel('K-value')
plt.ylabel('WCSS')
plt.show()
```

Final Model Building

```python
fin_cmodel=KMeans(n_clusters=3,random_state=42)
fin_cmodel.fit(scaled_crime)
```

```python
fin_cmodel.labels_
```
